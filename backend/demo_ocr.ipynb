```python
IMAGE_PATH = "../../samples/input/jml.png"

# # View Image
# from PIL import Image
# Image.open(IMAGE_PATH)

# View OCR Result
mistral_extractor = MistralFormExtractor()
parsed_form, ocr_text = mistral_extractor.structured_ocr(IMAGE_PATH)
print(parsed_form)
```
